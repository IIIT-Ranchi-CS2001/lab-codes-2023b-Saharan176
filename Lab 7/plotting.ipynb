```python
import matplotlib.pyplot as plt
```

```python
mp_party=['BJP','INC','BSP','others']
mp_seats=[163,66,10,20]
mp_explode=[0.1  if mp_seats[i] == max(mp_seats) else 0 for i in range(len(mp_party))]

raj_party=['INC','BJP','BSP','others']
raj_seats=[69,115,2,13]
raj_explode=[0.1  if raj_seats[i] == max(raj_seats) else 0 for i in range(len(raj_party))]
```

```python
plt.figure()
fig,(ax1,ax2)=plt.subplots(1,2,figsize=(10,5))
ax1.pie(mp_seats,labels=mp_party,autopct='%1.1f%%',startangle=90,explode=mp_explode,wedgeprops={'edgecolor':"black"})
ax1.set_title('M.P. Election Results')
ax2.pie(raj_seats,labels=raj_party,autopct='%1.1f%%',startangle=90,explode=raj_explode,wedgeprops={'edgecolor':"black"})
ax2.set_title('Rajasthan Election Results')
```

```python
mp_x=[i for i in range(len(mp_party))]
raj_x=[i+0.35 for i in range(len(mp_party))]
width=0.35
plt.bar(mp_x,mp_seats,width=width,label='Madhya Pradesh')
plt.bar(raj_x,raj_seats,width=width,label='Rajasthan')
plt.legend()
plt.xticks([i+0.175 for i in range(len(mp_party))],mp_party)
plt.ylabel('Seats Won')
plt.title("ASSEMBLY ELECTIONS")
```
