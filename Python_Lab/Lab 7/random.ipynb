```python
import random
```

```python
n,k = map(int,input("ENTER THE VALUES OF N AND K : ").split())
```

```python
def generateRandom(n,k):
    res=[]
    for i in range(k):
        res.append(random.randint(0,n))
    return res
```

```python
l=generateRandom(n,k)
```

```python
def is_prime(n):
    if n==1:
        return False 
    if n==2:
        return True 
    i=2
    while(i<=(n**(0.5)+1)):
        if n % i == 0:
            return False
        i+=1
    return True
```

```python
prime = list(filter(is_prime,l))
composite=list(filter(lambda x : not is_prime(x),l))
print(prime,composite)
```

```python
prime_square=list(map(lambda x:x**2,prime))
composite_root=list(map(lambda x:x**(0.5),composite))
print(prime_square,composite_root)
```

```python
import matplotlib.pyplot as plt
fig,(p,c)=plt.subplots(1,2,figsize=(10,5))
p.scatter(prime,prime_square)
p.set_xlabel('prime')
p.set_ylabel('prime_square')
c.scatter(composite,composite_root)
c.set_xlabel('composite')
c.set_ylabel('composite_root')
```
